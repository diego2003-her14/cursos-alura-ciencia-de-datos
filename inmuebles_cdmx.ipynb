## **Demandas de ML**

### ¿Cuáles son los valores promedio de alquiler por tipo de propiedad?

```python
datos.head()
```

```python
datos['Valor'].mean()
```

```python
datos.groupby('Tipo')['Valor'].mean()
```

```python
datos.groupby('Tipo')[['Valor']].mean().sort_values('Valor')
```

```python
df_tipo_precio = datos.groupby('Tipo')[['Valor']].mean().sort_values('Valor')

df_tipo_precio.plot(kind='barh', figsize=(12,8), color='purple')
```

```python
datos.Tipo.unique()
```

```python
inmuebles_comerciales = ['Conjunto Comercial/Sala', 'Edificio Completo', 'Tienda/Salón', 'Tienda en Centro Comercial'
                          'Casa Comercial', 'Terreno Estándar', 'Cochera/Estacionamiento', 'Hotel', 'Galpón/Depósito/Almacén', 'Loteo/Condominio', 'Industria']
```

```python
datos.query('@inmuebles_comerciales in Tipo')
```

```python
datos.query('@inmuebles_comerciales not in Tipo')
```

```python
df = datos.query('@inmuebles_comerciales not in Tipo')

df.head()
```

```python
df_tipo_precio = df.groupby('Tipo')[['Valor']].mean().sort_values('Valor')

df_tipo_precio.plot(kind='barh', figsize=(12,8), color='purple')
```

## Demandas de ultimo minuto

#####Calcular el promedio de habitaciones por departamento.

```python
C_promedio_habitaciones = df['Habitaciones'].mean()

C_promedio_habitaciones
```

#### Verificar cuántas colonias únicas existen en nuestra base de datos.

```python
colonias_unicas = df['Colonia'].unique()

colonias_unicas
```

#### Analizar qué colonias tienen el promedio de alquiler más alto.

```python
colonias_promedio = df.groupby('Colonia')[['Valor']].mean().sort_values('Valor', ascending=False)

colonias_promedio
```

#### Crear un gráfico de barras horizontales que muestre las 5 colonias con los promedios de alquiler más altos.

```python
colonias_promedio = df.groupby('Colonia')[['Valor']].mean().sort_values('Valor', ascending=False)

colonias_promedio.head(5).plot(kind='barh', figsize=(12,8), color='orange')
```
